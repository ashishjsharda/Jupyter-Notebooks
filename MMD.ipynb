```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the distribution
mean1 = [0, 0]
mean2 = [4, 4]
cov1 = [[1, 0], [0, 1]]
cov2 = [[1, 0], [0, 1]]

# Generate the samples from the distribution
num_samples = 100
samples1 = np.random.multivariate_normal(mean1, cov1, num_samples)
samples2 = np.random.multivariate_normal(mean2, cov2, num_samples)

# Calculate the MMD between the two sets of samples
def mmd(x, y, sigma=1):
    xx = np.dot(x, x.T)
    yy = np.dot(y, y.T)
    xy = np.dot(x, y.T)
    dist = np.mean(np.exp(-np.sum(xx, axis=1, keepdims=True) - np.sum(yy, axis=1, keepdims=True) + 2 * np.sum(xy, axis=1, keepdims=True) / sigma))
    return dist

mmd_value = mmd(samples1, samples2)

# Plot the samples and the MMD value
plt.scatter(samples1[:,0], samples1[:,1], c='blue', label='Samples1')
plt.scatter(samples2[:,0], samples2[:,1], c='red', label='Samples2')
plt.title('MMD Value: {:.2f}'.format(mmd_value))
plt.legend()
plt.show()
```
