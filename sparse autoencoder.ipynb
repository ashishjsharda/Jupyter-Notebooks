```python
import numpy as np
from keras.layers import Input, Dense
from keras.models import Model
from keras import regularizers

# Generate some random data
data = np.random.rand(1000, 100)

# Define the autoencoder architecture
input_layer = Input(shape=(100,))
hidden_layer = Dense(32, activation='relu', activity_regularizer=regularizers.l1(10e-5))(input_layer)
output_layer = Dense(100, activation='sigmoid')(hidden_layer)

# Create the autoencoder model
autoencoder = Model(input_layer, output_layer)

# Compile the model
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the autoencoder
autoencoder.fit(data, data, epochs=50, batch_size=32)

# Extract the learned compressed representation
encoder = Model(input_layer, hidden_layer)
compressed_data = encoder.predict(data)
```
