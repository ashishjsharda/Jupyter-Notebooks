```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 25, 20, 15]  # Represents the percentage distribution

# Create a pie chart
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)  # Create the pie chart with labels and percentages
```
